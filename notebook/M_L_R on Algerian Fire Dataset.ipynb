```python
df=pd.read_csv('Algerian_cleaned_dataset.csv')
```

```python
df=df.drop(columns=['day','month','year'])
```

```python
corr=df.corr()
corr
```

```python
df.head(2)
```

```python
x=df.drop(columns=['FWI'])
x
```

```python
y=df['FWI']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
x_train
```

```python
x_test.shape
```

```python
corr=x_train.corr()
corr
```

```python
plt.figure(figsize=(8,7))
sns.heatmap(corr,annot=True)
```

```python
def feature(dataset,thresold):
    col_corr=set()
    corr_matrix=x_train.corr()
    for i in range (len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>thresold:
                colname=corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr            
```

```python
drop_feature=feature(x_train,0.85)
drop_feature
```

```python
x_train.drop(drop_feature,axis=1,inplace=True)
x_test.drop(drop_feature,axis=1,inplace=True)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train_scaled=scaler.fit_transform(x_train)
x_train_scaled
```

```python
x_test_scaled=scaler.transform(x_test)
x_test_scaled
```

```python
plt.figure(figsize=(12,7))
plt.subplot(1,2,1)
sns.boxplot(x_train)
plt.title('x_train before scaled')

plt.subplot(1,2,2)
sns.boxplot(x_train_scaled)
plt.title('x_train after scaled')
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train_scaled,y_train)
```

```python
y_pred=model.predict(x_test_scaled)
y_pred
```

```python
plt.scatter(y_test,y_pred)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error,root_mean_squared_error
```

```python
r2=r2_score(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
rmse=np.sqrt(mse)
```

```python
print(f'r2:{r2}')
print(f'mse:{mse}')
print(f'mae:{mae}')
print(f'rmse:{rmse}')
```
